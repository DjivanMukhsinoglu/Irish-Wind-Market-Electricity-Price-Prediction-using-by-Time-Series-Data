```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Reading the updated data
df_updated = pd.read_csv('Day-ahead_Prices_2018-2023_BEST.csv')

# Convert the 'Time' column to datetime
df_updated['Time'] = pd.to_datetime(df_updated['Time'].str.split(' - ').str[0], format='%Y-%m-%d %H:%M:%S')

# Extract  11 AM
df_11_updated = df_updated[df_updated['Time'].dt.hour == 11]

# Set 'Time' as the index 
df_11_updated.set_index('Time', inplace=True)

# Convert  [EUR/MWh] to numeric
df_11_updated['Day-ahead Price [EUR/MWh]'] = pd.to_numeric(df_11_updated['Day-ahead Price [EUR/MWh]'], errors='coerce')

# Plot the day-ahead price at 11 AM
plt.figure(figsize=(15, 7))
plt.plot(df_11_updated['Day-ahead Price [EUR/MWh]'], label='Day-ahead Price at 11 AM')
plt.legend(loc='best')
plt.title('Day-ahead Price at 11 AM Time Series')
plt.ylabel('Price [EUR/MWh]')
plt.xlabel('Time')
plt.grid(True)
plt.show()
```
