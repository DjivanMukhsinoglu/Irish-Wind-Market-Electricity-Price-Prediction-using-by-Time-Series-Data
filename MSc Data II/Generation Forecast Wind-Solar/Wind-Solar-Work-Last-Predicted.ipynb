```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('complete_generation_forecasts_with_lstm_predictions.csv')

# Convert the 'Time' column to datetime format
df['Time'] = pd.to_datetime(df['Time'])

# Set 'Time' as the index
df.set_index('Time', inplace=True)

# Resample the data to daily frequency
df_daily = df.resample('D').mean()

# Visualize the daily time series data 
df_daily.plot(figsize=(12, 6))
plt.xlabel("Time")
plt.ylabel("Wind Offshore [MW] Day Ahead")
plt.title("Daily Time Series of Wind Offshore [MW] Day Ahead")
plt.grid(True)
plt.show()
```
